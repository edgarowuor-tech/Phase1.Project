DATA MANIPULATION

```python
#changing the Event date to datetime format
aviation_df['Event.Date'] = pd.to_datetime(aviation_df['Event.Date'])
```

```python
#changing the Publishing date to datetime format
aviation_df['Publication.Date'] = pd.to_datetime(aviation_df['Publication.Date'])
```

```python
aviation_df.info()
```

```python
#column 'Weather Condition' has categorical data UNK, VMC, IMC,Unk ,i want to standardize the Unk to UNK
aviation_df['Weather.Condition'] = aviation_df['Weather.Condition'].replace('Unk', 'UNK')
aviation_df['Weather.Condition'].value_counts()
```

```python
#Standardizing column names by removing full stop and capitalizing each word
aviation_df.columns = aviation_df.columns.str.replace(".", " ", regex=False).str.title()
aviation_df.columns
```

```python
aviation_df
```

DATA CLEANING

```python
aviation_df.isnull().sum()
```

```python
#checking the percentage of missing values in each column
missing_percentage = (aviation_df.isnull().sum() / len(aviation_df))*100
missing_percentage
```

Dropping columns ,Latitude ,Longitude,Aircraft Category,Schedule,Far Description,Air Carrier.It is difficult to draw insights from these columns since a significant percentage of the values are missing

```python
#dropping columns with a significant percentage of missing values
aviation_df = aviation_df.drop(columns=['Latitude', 'Longitude', 'Aircraft Category', 'Schedule', 'Far Description', 'Air Carrier'])
aviation_df.info()
```

```python
#removing rows with missing values in Publication Date,Amateur Built,Location,Country,Make ,Registration Number ,Injury Severity and Model columns
aviation_df = aviation_df.dropna(subset=['Publication Date', 'Amateur Built', 'Location','Country', 'Make', 'Model', 'Registration Number','Injury Severity'])
aviation_df.isnull().sum()
```

CHECKING FOR DUPLICATES

```python
aviation_df.duplicated().sum()
```

```python
aviation_df.isnull().sum()
```

```python
#Filling missing values in these columns with unknown rather than deleting them
aviation_df[["Engine Type", "Number Of Engines", "Purpose Of Flight"]] = aviation_df[["Engine Type", "Number Of Engines", "Purpose Of Flight"]].fillna("Unknown")
aviation_df.isnull().sum()
```

```python
aviation_df["Report Status"]= aviation_df["Report Status"].fillna("Unknown")
```

```python
aviation_df["Weather Condition"]= aviation_df["Weather Condition"].fillna("Others")
```

```python
aviation_df["Aircraft Damage"]= aviation_df["Aircraft Damage"].fillna("Unknown")
```

```python
aviation_df.isnull().sum()
```

```python
#Fill with 0 (since no record of injuries often means none reported
aviation_df["Total Fatal Injuries"] = aviation_df["Total Fatal Injuries"].fillna(0)
aviation_df["Total Serious Injuries"] = aviation_df["Total Serious Injuries"].fillna(0)
aviation_df["Total Minor Injuries"] = aviation_df["Total Minor Injuries"].fillna(0)
aviation_df["Total Uninjured"] = aviation_df["Total Uninjured"].fillna(0)
```

```python
aviation_df['Broad Phase Of Flight'].value_counts()
```

```python
aviation_df.isnull().sum()
```

```python
aviation_df["Broad Phase Of Flight"]= aviation_df["Broad Phase Of Flight"].fillna("Unknown")
```

```python
aviation_df[["Airport Code", "Airport Name"]]=aviation_df[["Airport Code", "Airport Name"]].fillna('Unknown')
aviation_df.info()
```

DATA VISUALIZATION USING MATPLOTLIB AND SEABORN

```python
# Accidents by Broad Phase of Flight
aviation_df.groupby("Broad Phase Of Flight").size().sort_values(ascending=False).plot(kind="bar")
plt.title("Accidents by Phase of Flight")
plt.xlabel("Phase of Flight")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```

 Flight phase: Excluding the “Unknown” category, the Landing phase record the highest number of accidents followed by Take Off.
 
 Recommendations:Pilot Training: focus on takeoff & landing safety.

```python
#Accidents by Weather Condition
aviation_df.groupby("Weather Condition").size().sort_values(ascending=False).plot(kind="bar")
plt.title("Accidents by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.show()
```


 VMC=Visual Meteorological Conditions (This means the weather allowed pilots to see outside and navigate visually)
 
IMC=Instrument Meteorological Conditions(This means poor visibility i.e fog,clouds)

Weather:Most accidents occurred under Visual Meteorological Conditions (VMC), reflecting that while good weather enables more flights, it also exposes a greater share of accidents compared to IMC

Recommendation:Strengthen pilot training and safety protocols in VMC operations, since higher traffic in good weather increases the likelihood of accidents despite favorable conditions

```python
#Top Aircraft Manufacturers in Accidents
aviation_df.groupby("Make").size().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Aircraft Manufacturers with Accidents")
plt.xlabel("Aircraft Make")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```

Cessna and Piper had so a high number of accidents

Recomendation
Enhance maintenance oversight and pilot training programs for Cessna and Piper aircraft, as their high accident numbers suggest the need for stricter safety checks and operational best practices for these widely used models.

```python
# Group by Year
year = aviation_df.groupby(aviation_df["Event Date"].dt.year).size().reset_index(name="Accidents")

# Plot trend
plt.figure(figsize=(10,6))
sns.lineplot(data=year, x="Event Date", y="Accidents", marker="o", color="purple")
plt.title("Trend of Aviation Accidents Over Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

Trends over time: overall accidents decreased, but spikes remain.

Recommendation
Overall accident rates have shown a downward trend over time, but periodic spikes highlight the need for continuous monitoring and targeted safety interventions during high-risk periods.
